Exploratory Data Analyst

```python
#Menampilkan distribusi beberapa kolom
df['Branch'].value_counts().plot(kind='bar', title='Branch Distribution')
plt.show()

df['City'].value_counts().plot(kind='bar', title='City Distribution')
plt.show()

df['Customer type'].value_counts().plot(kind='bar', title='Customer Type Distribution')
plt.show()

df['Product line'].value_counts().plot(kind='bar', title='Product Line Distribution')
plt.show()

df['Payment'].value_counts().plot(kind='bar', title='Payment Method Distribution')
plt.show()
```

```python
#Visualisasi distribusi rating
plt.figure(figsize=(10,6))
sns.histplot(df['Rating'], bins=10, kde=True)
plt.title('Distribution of Customer Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.set(style="whitegrid")
```

Distribusi Rating

```python
plt.figure(figsize=(10, 6))
colors = sns.color_palette("husl", 10)
n, bins, patches = plt.hist(df['Rating'], bins=10, edgecolor='black')

# Menggunakan warna yang berbeda untuk setiap bar
for i in range(len(patches)):
    patches[i].set_facecolor(colors[i % len(colors)])

plt.title('Distribution of Customer Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Total Penjualan Per Branch

```python
plt.figure(figsize=(10, 6))
colors = sns.color_palette("Set2", 3)
sns.barplot(x='Branch', y='Total', data=df, estimator=np.sum, palette=colors)
plt.title('Total Sales per Branch')
plt.xlabel('Branch')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

Total Penjualan Berdasarkan Product Line

```python
plt.figure(figsize=(12, 8))
colors = sns.color_palette("Paired", len(df['Product line'].unique()))
sns.barplot(x='Product line', y='Total', data=df, estimator=np.sum, palette=colors)
plt.title('Total Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Konversi kolom 'date' ke datetime
df['Date'] = pd.to_datetime(df['Date'])

# Tambahkan kolom 'day' yang menunjukkan hari dalam seminggu
df['day'] = df['Date'].dt.day_name()

# Cek hasilnya
df[['Date', 'day']].head()
```

```python
df_per_day=df.groupby('day')['Total'].sum().reindex(['Monday','Tuesday''Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.figure(figsize=(10, 6))
sns.lineplot(x=df_per_day.index, y=df_per_day.values, marker='o', color='b')
plt.title('Total Sales per Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales')
plt.show()
```

Penjualan Berdasarkan Hari dalam Seminggu

Penjualan Berdasarkan Jam dalam Sehari

```python
# Konversi kolom 'time' ke datetime dan ekstrak jam
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M')
df['hour'] = df['Time'].dt.hour

# Cek hasilnya
df[['Time', 'hour']].head()
```

```python
df_per_hour = df.groupby('hour')['Total'].sum()
plt.figure(figsize=(10, 6))
sns.lineplot(x=df_per_hour.index, y=df_per_hour.values, marker='o', color='g')
plt.title('Total Sales per Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Mengatur ukuran plot
plt.figure(figsize=(10, 6))

# Menggunakan palet warna 'husl' dari seaborn dengan 3 warna (sesuai jumlah cabang)
colors = sns.color_palette("husl", 3)

# Membuat count plot untuk cabang (Branch) dengan warna yang berbeda
sns.countplot(x='Branch', data=df, palette=colors, hue='Branch', legend=False, edgecolor='black')

# Menyertakan label sumbu dan judul
plt.xlabel('Branch Name')
plt.ylabel('Count')
plt.title('Which Branch is the most busy?')
A,B,C =df.Branch.value_counts()

print('Branch A -',A)
print('Branch B -',C)
print('Branch C -',B)

# Menampilkan plot
plt.show()
```
